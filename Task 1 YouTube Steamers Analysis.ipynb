```python
youtubers_country_count = df['Country'].value_counts().head(15)
youtubers_country_count
```

```python
youtubers_country_count.plot(kind = 'bar', title = 'Distribution of Youtube Streamers across Top 15 Countries')
plt.xlabel("Country")
plt.ylabel("Number of Youtubers")
plt.show()
```

```python
top_categories_count = df['Categories'].value_counts()
top_categories_count
```

```python
top_categories_count.plot(kind = 'bar', title = 'Youtube Categories')
plt.xlabel("Categories")
plt.ylabel("Number of Youtube Streamers")
plt.show()
```
